```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#rename columns
```

```python
#Year of event	int64
#Event dates	object
#Event name	object
#Event distance/length	object
#Event number of finishers	int64
#Athlete performance	object
#Athlete gender	object
#Athlete average speed	float64
#Athlete ID	int64
#athlete_age	int64
```

```python
df2 = df2.rename(columns = {'Year of event':'year',
                            'Event dates':'race_day',
                            'Event name':'race_name',
                            'Event distance/length':'race_length',
                            'Event number of finishers':'race_number_of_finishers',
                            'Athlete performance':'athlete_performance',
                            'Athlete gender':'athlete_gender',
                            'Athlete average speed':'athlete_average_speed',
                            'Athlete ID':'athlete_id'
})
```

```python
df2.head()
```

```python
# reorder columns and not include year
```

```python
df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
# Find the difference in speed for the 50km,50mi for male and female
```

```python
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# Find what age groups are best in the 50m race (Min. of 20+ races)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['count','mean']).sort_values('mean', ascending = False).query('count >= 20').head(15)
```

```python
# Find what age groups are the worst in the 50mi race (Min. of 20+ races)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['count','mean']).sort_values('mean', ascending = True).query('count >= 20').head(15)
```

```python
# Adding Seasons for the data, are runners slower in summer compared to the winter?
# Spring 3-5
#Summer 6-8
# Fall 9-11
# Winter 12-2
# Split between two decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(20)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x in [12,1,2] else 'Fall' if x in [9,10,11] else 'Summer' if x in [6,7,8] else 'Spring' if x in [2,3,4] else 'NA')
```

```python
df3.head(50)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
#50 Miler Only
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
